```python
lst_company = ["NOVN.SW", "AAPL", "PFE"]

for comp in lst_company:
    company = yf.Ticker(comp).info
    ticker = company["underlyingSymbol"]
    longName = company["longName"]
    sector = company["sector"]

    print(ticker, longName, sector)
```

```python
for comp in lst_company:
    company = yf.Ticker(comp).info
    ticker = company["underlyingSymbol"]
    country = company["country"]
    city = company["city"]
    industry = company["industry"]
    marketCap = company["marketCap"]

    print(ticker, country, city, industry, marketCap)
```
